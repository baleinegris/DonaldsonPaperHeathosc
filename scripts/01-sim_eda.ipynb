```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("../data/00-simulated_data/simulated_data.csv")
```

```python
# plot number of buildings by year built
fig = px.histogram(df, x="year_built", nbins=25, title="Distribution of Year Built")
fig.show()
```

```python
# plot average number of stories by year built
fig = px.histogram(df, x="year_built", y="storeys", histfunc="avg", title="Average Number of Stories by Year Built")
fig.show()
```

```python
# plot average number of units by year built
fig = px.histogram(df, x="year_built", y="units", histfunc="avg", title="Average Number of Units by Year Built")
fig.show()
```

```python
# plot total number of units by year built
fig = px.histogram(df, x="year_built", y="units", nbins=25, title="Total Units by Year Built", histfunc="sum")
fig.show()
```
